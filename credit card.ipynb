```python
def check_card(card_number):
    digits = [int(d) for d in str(card_number)]
    
    for i in range(len(digits)-2, -1, -2):
        digits[i] = digits[i] * 2
        if digits[i] > 9:
            digits[i] -= 9

    if sum(digits) % 10 == 0:
        return "Valid Credit Card"
    else:
        return "Invalid Credit Card"

card = input("Enter card number: ")
print(check_card(card))
```
